### Conslusion

1. Neural Networks with peer prediction loss have the best overall performance, and outperforms regular neural networks on almost all tasks.
1. The neural networks have more potential because I used the same parameters for all tasks. The performance could be further improved by tuning parameters.
1. NN with peer prediction loss is very robust to perturbation, which is why it performs relevantly better on harder tasks.
