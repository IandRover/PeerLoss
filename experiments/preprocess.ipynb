```python
import pandas as pd
import numpy as np

def load(path):
    df = open(path).readlines()
    df = list(map(lambda line: list(map(float, line.split())), df))
    df = pd.DataFrame(df)
    return df

def preprocess(name):
    train_data = load(f'uci/{name}/{name}_train_data_1.asc').values
    train_labels = load(f'uci/{name}/{name}_train_labels_1.asc').values
    test_data = load(f'uci/{name}/{name}_test_data_1.asc').values
    test_labels = load(f'uci/{name}/{name}_test_labels_1.asc').values
    
    data = np.concatenate([train_data, test_data])
    labels = np.concatenate([train_labels, test_labels])
    
    dataset = np.concatenate([labels, data], axis=1)
    df = pd.DataFrame(dataset)
    df.rename(columns={0: 'target'}, inplace=True)
    df.target.replace({-1: 0}, inplace=True)
    df.to_csv(f'uci/{name}.csv', index=False)
    return df
```

```python
preprocess('image')
```

```python
preprocess('titanic')
```
